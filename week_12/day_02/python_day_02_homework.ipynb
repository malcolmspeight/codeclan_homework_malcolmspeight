## Question 3

```python
# subset columns
drinks = drinks_content.loc[:, ["Beverage_category", "Beverage",
                               "Beverage_prep", "Calories"]]
drinks
```

```python
# missing values
drinks.isnull().any()
```

## Question 4

```python
# create new column
drinks["calorie_diff"] = drinks["Calories"] - 135
drinks.head(5)
```

## Question 5

```python
# average calories per beverage category
drinks.groupby("Beverage_category")["Calories"].mean()
```

## Question 6

```python
# average calories across all drinks
avg_calories = drinks["Calories"].mean()

avg_calories
```

```python
# drinks dataset summarised by bev_prep with avg calories
bev_prep_drinks = drinks.groupby("Beverage_prep").mean()[["Calories"]]

bev_prep_drinks
```

```python
# bev_prep drinks with avg cals > average calories
bev_prep_drinks.loc[bev_prep_drinks["Calories"] > avg_calories]
```
